```python
import pandas as pd
import plotly.express as px

olympic_data = pd.read_csv("Data/athlete_events.csv")
olympic_data.head()
```

#### Sports
* 1. Basketball
* 2. Football
* 3. Ice Hockey

#### medaljfördelning mellan länder i sporterna

```python
sports = ['Basketball', 'Football', 'Ice Hockey']

# Loop through each of the sports we selected and create graphs
for sport in sports:
    # drop all rows that dont involve our sport
    olympic_data_sport = olympic_data.drop(olympic_data[olympic_data['Sport'] != sport].index)
    
    # get all medals (remove nan's)
    olympic_data_sport_with_medals = olympic_data_sport[olympic_data_sport['Medal'].notna()]
    
    # Group by 'NOC' and count all medals
    medals_per_country_sport = olympic_data_sport_with_medals.groupby('NOC')['Medal'].count()

    medals_per_country_sport = medals_per_country_sport.sort_values(ascending=False)

    fig = px.bar(
        x=medals_per_country_sport.index,
        y=medals_per_country_sport,
        title=f"Medal distribution across countries in {sport.lower()}",
        labels={
            'x': '',
            'y': 'Medals'
        }
    )

    fig.show()
```

#### åldersfördelning i sporterna

```python
# Get row indexes that dont have our 
olympic_data_sports = olympic_data.query("Sport == 'Basketball' | Sport == 'Football' | Sport == 'Ice Hockey'")

fig = px.histogram(
    data_frame=olympic_data_sports,
    color='Sport',
    x="Age",
    title="Age distribution across sports"
)

fig.show()
```

#### Which sport has the tallest athletes?

```python
olympic_data_sports_height = olympic_data_sports.query("Height > 195")

fig = px.histogram(
    data_frame=olympic_data_sports_height,
    color='Sport',
    x="Sport",
    y = "Height",
    title="Height distribution across sports"
)

fig.show()
```

#### Which sport has the heaviest athletes?

```python
olympic_data_sports_weight = olympic_data_sports.query("Weight > 110")

fig = px.histogram(
    data_frame=olympic_data_sports_weight,
    color='Sport',
    x="Sport",
    y = "Weight",
    title="Which sports have the heaviest athletes?"
)

fig.show()
```
